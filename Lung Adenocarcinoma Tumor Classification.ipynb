## Dataset

This project uses the **TCGA-LUAD STAR TPM gene expression dataset**.  
LUAD stands for **Lung Adenocarcinoma**, which is a subtype of lung cancer.

The dataset contains gene expression profiles from TCGA samples and is used for a binary classification task:

- **Tumor**: primary lung adenocarcinoma tumor samples
- **Normal**: solid tissue normal samples

## Project Target

The main objective of this project is to classify lung adenocarcinoma samples into:

- **Tumor**
- **Normal**

using machine learning based on gene expression data.

## Load File

```python
# 3. Define all file paths based on PROJECT_DIR
RAW_FILE = PROJECT_DIR / "TCGA-LUAD.star_tpm.tsv.gz"
PROCESSED_FILE = PROJECT_DIR / "tcga_luad_star_tpm_ml_dataset.parquet"

LOCAL_DIR = Path("/content/tcga_luad")
LOCAL_DIR.mkdir(parents=True, exist_ok=True)

LOCAL_FILE = LOCAL_DIR / RAW_FILE.name

# 4. Validate raw file exists
if not RAW_FILE.exists(): raise FileNotFoundError(f"File not found:\n{RAW_FILE}\n\n")

# 5. Copy raw file to Colab local runtime for faster reading
if not LOCAL_FILE.exists():
    shutil.copy2(RAW_FILE, LOCAL_FILE)

print(f"Local file ready: {LOCAL_FILE}")

# 6. Load expression matrix
expr_raw = pd.read_csv(LOCAL_FILE, sep="\t", compression="infer")

# print("Raw expression matrix shape:", expr_raw.shape)
# display(expr_raw.head())
```

# Raw File Processing

```python
# 7. Detect gene identifier column
gene_col = expr_raw.columns[0]

# 8. Convert gene x sample into sample x gene
expr_ml = expr_raw.set_index(gene_col).T.reset_index()
expr_ml = expr_ml.rename(columns={"index": "sample_id"})

# 9. Create Tumor/Normal label from TCGA barcode
def get_label_from_tcga_barcode(barcode):
    parts = str(barcode).split("-")

    if len(parts) < 4:
        return "Unknown"

    sample_type_code = parts[3][:2]

    if sample_type_code in ["01", "02", "03", "04", "05", "06", "07", "08", "09"]:
        return "Tumor"
    elif sample_type_code in ["10", "11", "12", "13", "14"]:
        return "Normal"
    else:
        return "Unknown"

expr_ml["label"] = expr_ml["sample_id"].apply(get_label_from_tcga_barcode)

# 10. Keep only Tumor and Normal samples
expr_ml = expr_ml[expr_ml["label"].isin(["Tumor", "Normal"])].reset_index(drop=True)

# 11. Save processed dataset back to Google Drive
expr_ml.to_parquet(PROCESSED_FILE, index=False)

# 12. Display processed data only
display(expr_ml.head())
```
